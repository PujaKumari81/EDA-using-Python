```python
#Movies runtime greater than 200 min
```

```python
imdb.columns
```

```python
imdb['Runtime']=imdb['Runtime'].str.replace('min','')
```

```python
imdb['Runtime']=imdb['Runtime'].apply(int)
```

```python
imdb1=imdb[imdb['Runtime']>=200][['Series_Title','Runtime']].set_index('Series_Title').head(5)
```

```python
imdb1
```

```python
fig=plt.figure(figsize=(15,4))
sns.barplot(x=imdb1.index,y='Runtime',data=imdb1)
```

```python
#in which year was the highest voting made
```

```python
data1=imdb.groupby('Released_Year')['No_of_Votes'].mean().sort_values(ascending=False).head(5)
data1
```

```python
#Top 10 Highest Gross Movie Titles
```

```python
emo=imdb.nlargest(10,'Gross')[['Series_Title','Gross']].set_index('Series_Title')
```

```python
emo
```

```python
sns.barplot(x='Gross',y=emo.index,data=emo)
plt.title("Top 10 Highest Revenue Movie Titles")
```

```python
dat1=sum(imdb['Genre'].str.contains('Drama',case=False))
```

```python
dat1
```

```python
imdb['Genre']
```

```python
dat2=[]
```

```python
for value in imdb['Genre']:
    dat2.append(value.split(','))
```

```python
dat2
```

```python
dat3=[]
    
```

```python
for item in dat2:
    for item1 in item:
        dat3.append(item1)
```

```python
dat3.strip()
```

```python
dupes=[]
```

```python
for item in dat3:
    if item not in dupes:
        dupes.append(item)
```
